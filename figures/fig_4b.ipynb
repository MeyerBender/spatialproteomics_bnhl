# Legends

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def create_legend(label_color_dict, fontsize=6):
    """
    Creates a legend from a dictionary of labels and colors with rectangles.

    Parameters:
    - label_color_dict (dict): A dictionary where keys are labels and values are colors.
    - fontsize (int): Font size for the legend text.

    Returns:
    - A legend object.
    """
    legend_elements = [
        mpatches.Patch(color=color, label=label) for label, color in label_color_dict.items()
    ]
    
    legend = plt.legend(
        handles=legend_elements, 
        frameon=False,  # Turn off the frame
        fontsize=fontsize
    )
    
    return legend

# Example usage
label_color_dict = {
    "B cell": "#5799d1",
    "T cell": "#ebc850",
    "Myeloid cell": "#de6866"
}

plt.figure(figsize=(1, 1), dpi=900)
create_legend(celltype_colors)
plt.axis("off")  # Hide axes
plt.show()
```
